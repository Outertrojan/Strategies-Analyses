```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


def binomial_random_walk(S0, u, d, p, n_steps):
    """
    Function to simulate a binomial random walk 

    Parameters:
    S0 (float): Initial stock price.
    u (float): Factor of price increase on upward movement.
    d (float): Factor of price decrease on downward movement.
    p (float): Probability of an upward movement.
    n_steps (int): Number of steps in the random walk.

    Returns:
    np.array: Array containing the simulated stock price path.
    """
    # Generate random steps: 1 for up, 0 for down
    steps = np.random.choice([1, 0], size=n_steps, p=[p, 1-p])
    # print(steps)

    # Calculate price movements
    price_movements = u ** steps * d ** (1 - steps)
    # print(price_movements)
    # print(1-steps)
    # Calculate stock price path
    price_path = S0 * np.cumprod(price_movements)

    # Insert initial stock price at the beginning of the price path
    price_path = np.insert(price_path, 0, S0)

    return price_path


# Example usage
S0 = 100  # Initial stock price
u = 1.1   # Upward movement factor
d = 0.9   # Downward movement factor
p = 0.5   # Probability of upward movement
n_steps = 1000  # Number of steps

# Generate a binomial random walk
price_path = binomial_random_walk(S0, u, d, p, n_steps)

# Print the simulated stock price path
print("Simulated Stock Price Path")

sns.set()
plt.figure(figsize=(10, 8))
plt.plot(price_path)
plt.show()
```
